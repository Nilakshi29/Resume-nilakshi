```python
#Q2-How many rows and columns are there?

#number of rows
row=len(df.index)
#no of columns
col= len(df.columns)

print("no of rows=", row, "and no of columns=", col)
```

```python
#Q3:What is the average Price of all cars in the dataset?

avg= df['price'].mean()
print('Average priceof cars=', avg)
```

```python
#Q4:Which is the cheapest make and costliest make of car in the lot?

a= df[df['price']==df.max]

b= df[df['price']== df.min]

print('Max value: ', a, 'Min value: ', b)
```

```python
#Q5:How many cars have horsepower greater than 100?

num=len(df[df['horsepower']>100])

print(num)
```

```python
#Q6: How many hatchback cars are in the dataset ?

len(df[df['body_style']=='hatchback'])
```

```python
#*Q7: What are the 3 most commonly found cars in the dataset?

#We want most common thats why using head as 3
n=df['make'].value_counts().head(3)
print('No of unique cars', n)
```

```python
#Q8:Someone purchased a car for 7099, what is the make of the car?

making=(df['price']==7099).info('make')
print('MAke of the car is', making)
```

```python
#Q9: Which cars are priced greater than 40000?

print('cars priced greater than 40000')
df[df['price']>40000]
```

```python
#10: Which are the cars that are both a sedan and priced less than 7000?

print('Following are the cars having both Sedan as well as Price less thyan 7000')
df[(df['price']<7000) & (df['body_style']=='sedan')]
```
